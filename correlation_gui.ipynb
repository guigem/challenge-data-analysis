```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/immoweb_cleaned.csv")
data = data.set_index("id")
data.head(10)
```

```python
data.shape
```

```python
Wallonia = data.loc[data['region'] == "WAL"]
Flandern = data.loc[data['region'] == "VLA"]
Brussels = data.loc[data['region'] == "BXL"]
Wallonia.head(10)
```

## Belgium

```python
corrMatrix = data.corr()
corrMatrix
```

```python
fig_dims = (12, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.heatmap(corrMatrix, annot=True, ax = ax)
plt.show()
```

```python
# Boxplot of the price for each province
fig_dims = (12, 10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.boxplot(x='Province',
               y='price', 
               data=data,
               ax = ax) 
plt.title("Boxplot of the price for each province")
plt.xlabel("Province")
plt.ylabel("Price")
               
```

```python
data.groupby('locality')['price'].mean().idxmax()
#9772 is Wannegem-Lede
```

```python
data.groupby('locality')['price'].mean().idxmin()
#7610 Rumes 
```
