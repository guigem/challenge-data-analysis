## Importing the libraries

```python
import pandas as pd
import numpy as np
```

## Reading the csv + putting "id" as index

```python
data = pd.read_csv("data/immoweb_scrapped.csv")
data = data.set_index("id")
data.head(10)
```

## Checking if the index is unique

```python
data.index.is_unique
```

## Removing rows without price

```python
data = data.dropna(subset=['price'])
data.head(10)
```

## Filling missing values : 
 - Swimming_pool with False
 - Nr_of_facades with -1

```python
data['swimming_pool'] = data['swimming_pool'].fillna(False)
data['nr_of_facades'] = data['nr_of_facades'].fillna(-1)
data.head(75)
```

## Checking the types of columns

```python
data.dtypes
```

## Change type of locality, nr_of_rooms, furnished, terrace, garden, nr_of_facades, swimming_pool

```python
convert_dict = {'nr_of_rooms': int,
                'furnished': bool,
                'terrace': bool,
                'garden': bool, 
                'nr_of_facades': int,
                'swimming_pool': bool 
               } 
  
data = data.astype(convert_dict) 
```

```python
data.dtypes
```

## Adding province column 

```python
#function to recognize the province based on the postal code
def province(x):
    
    prov = []
    for i in x["locality"]:
        
        if 2000 <= i <= 2999:
            
            #Antwerp
            prov.append("VAN")

        if 3500 <= i <= 3999:
            
            #Limburg
            prov.append("VLI")
            
        if 4000 <= i <= 4999:
            
            #Liège
            prov.append("WLG")
            
        if 1300 <= i <= 1499:
            
            #Brabant Wallon
            prov.append("WBR")
            
        if 5000 <= i <= 5999:
            
            #Namur
            prov.append("WNA") 
            
        if 6600 <= i <= 6999:
            
            #Luxembourg
            prov.append("WLX")
            
        if 1000 <= i <= 1299:
            
            #Brussels
            prov.append("Brussels")
              
 
        if (1500 <= i <= 1999) | (3000 <= i <= 3499):
            
            #Brabant Flamand
            prov.append("VBR")


        if (6000 <= i <= 6599) | (7000 <= i <= 7999):
            
            #Hainaut
            prov.append("WHT")
            
        if 8000 <= i <= 8999:
            
            #Flander Occidentale
            prov.append("VWV")   
            
        if 9000 <= i <= 9999:
            
            #Flander Orientale
            prov.append("VOV")   
            
    return prov
```

```python
data["Province"] = province(data)
data.head(10)
```
